```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated FACS data
protein_expression = np.random.normal(loc=100, scale=20, size=1000)

# Histogram of protein expression levels
plt.hist(protein_expression, bins=50, color='blue', alpha=0.7)
plt.title('Protein Expression Levels')
plt.xlabel('Fluorescence Intensity')
plt.ylabel('Cell Count')
plt.show()
```
